```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = [[59,765],
        [146,20]]

columns = ('Detected by YOLO','Undetected by YOLO')
rows = ["Data labeled abnormal", "Data labeled normal"]

values = np.arange(0, 1200, 200)
value_increment = 1
```

```python
# Get some pastel shades for the colors
colors = plt.cm.BuPu(np.linspace(0.3, 0.5, len(rows)))
n_rows = len(data)

index = np.arange(len(columns)) + 0.3
bar_width = 0.7

# Initialize the vertical-offset for the stacked bar chart.
y_offset = np.zeros(len(columns))

# Plot bars and create text labels for the table
cell_text = []
for row in range(n_rows):
    plt.bar(index, data[row], bar_width, bottom=y_offset, color=colors[row])
    y_offset =data[row]
    cell_text.append(['%d' % x for x in y_offset])
# Reverse colors and text labels to display the last value at the top.
colors = colors[::-1]
cell_text.reverse()

# Add a table at the bottom of the axes
the_table = plt.table(cellText=cell_text,
                      rowLabels=rows,
                      rowColours=colors,
                      colLabels=columns,
                      loc='bottom')

# Adjust layout to make room for the table:
plt.subplots_adjust(left=0.3, bottom=0.01)

plt.ylabel('amount of data')
plt.yticks(values * value_increment, ['%d' % val for val in values])
plt.xticks([])
plt.title("Prediction result of YOLO")
plt.savefig('new_YOLO_result.pdf',dpi=1000,bbox_inches="tight")
plt.show()
```

```python
plt.pie([146,59], labels = ["Data labeled abnormal","Data labeled normal"], autopct = '%1.1f%%')
plt.title('Distribution of data detected as abnormal by YOLO')
plt.axis('equal')
plt.legend(loc='best')
plt.savefig('new_YOLO_result_detected.pdf',dpi=100,bbox_inches="tight")
```
