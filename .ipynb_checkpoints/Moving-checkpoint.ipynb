```python
cap = cv2.VideoCapture(path_single_video)
timeF = 50
c = 0
# 以迴圈從影片檔案讀取影格，並顯示出來
while(cap.isOpened()):
    ret, frame = cap.read()    
    if frame is None:
        break
    if(c%timeF==0):
        #frame = frame[100:250,350:500]
        #frame = frame[10:,360:]
        if c == 0:
            previous = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
            previous = cv2.GaussianBlur(previous,(7,7),0)
            #ret,previous = cv2.threshold(frame,200,255,cv2.THRESH_BINARY)
            
        else:
            current = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
            current = cv2.GaussianBlur(current,(7,7),0)
            #ret,current = cv2.threshold(frame,200,255,cv2.THRESH_BINARY)
#             print('current:')
#             print(current[0:5,0:5])
#             print('previous:')
#             print(previous[0:5,0:5])
            minus = cv2.absdiff(current,previous)
            ret,minus = cv2.threshold(minus,10,255,cv2.THRESH_BINARY)
            #cv2.imwrite("../FullVideo/20180628/"+str(c)+".jpg",minus)
            #minus = current - previous
            #minus = np.absolute(minus)
#             print('minus:')
#             print(minus[0:5,0:5])
            previous = current
            #cv2.imshow('mytitle',minus)
            cv2.imshow('mytitle',minus)
            print(np.sum(minus/255))
            if cv2.waitKey(1) & 0xFF == ord('q'):
                break
    c+=1
    if c == 10000:
        break
    
cap.release()
cv2.destroyAllWindows()
```
